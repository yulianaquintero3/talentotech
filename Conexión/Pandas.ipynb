```python
df['imc2'] = df['peso'] / (df['altura']**2)
df
```

```python
def categorize_imc(imc):
  if imc < 18.5:
    return 'bajo de peso'
  elif imc < 25:
    return 'peso normal'
  else:
    return 'sobrepeso'

df['categora'] = df['imc'].apply(categorize_imc)
display(df)
```
