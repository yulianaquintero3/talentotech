```python
import math

n1 = float(input("Enter the 1st number: "))
n2 = float(input("Enter the 2nd number: "))

sum = lambda a,b: a+b
result = sum(n1, n2)
print(f"The sum is {result}")

substraction = lambda a,b: a-b
result = substraction(n1, n2)
print(f"The substraction is {result}")

multiplication = lambda a,b: a*b
result = multiplication(n1, n2)
print(f"The multiplication is {result}")

division = lambda a,b: a/b
result = division(n1, n2) if n2!=0 else "You cannot divide by zero"
print(f"The division is {result}")

power = lambda base,exponente: math.pow(base,exponente)# base**exponente
result = power(n1, n2)
print(f"The power is {result}")

root  = lambda a,b: a**(1/b)
result = root(n1,n2)
print(f"The root is {result}")

logarithm = lambda a,b: math.log(a,b)
result = logarithm(n1,n2)
print(f"The logarithm is {result}")

sine = lambda angulo_radianes: math.sin(angulo_radianes)
result = sine(n1)
print(f"The sine is {result}")

cosine = lambda angulo_radianes: math.cos(angulo_radianes)
result = cosine(n1)
print(f"The cosine is {result}")

tangent = lambda angulo_radianes: math.tan(angulo_radianes)
result = tangent(n1)
print(f"The tangent is {result}")
```
