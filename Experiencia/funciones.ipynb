```python
import math
def datos():
  numero_uno=float(input('Ingrese el valor del cateto #1: '))
  numero_dos=float(input('Ingrese el valor del cateto #2: '))
  return numero_uno,numero_dos

def calcular_hipotenusa(numero_uno,numero_dos):
  hipotenusa=math.sqrt(numero_uno**2 + numero_dos**2)
  print("La hipotenusa es: ", hipotenusa)

def llamarfuncion():
  numero_uno,numero_dos=datos()
  calcular_hipotenusa(numero_uno,numero_dos)

llamarfuncion()
```

```python

def datos():
  hora=float(input('Ingrese la hora: '))
  return hora

def saludar_segun_hora(hora):
    hora_actual = hora
    if 0 <= hora_actual < 13:
        print("¡Buenos días!")
    if 13 <= hora_actual < 19:
        print("¡Buenas tardes!")
    if 19 <= hora_actual < 24:
        print("¡Buenas noches!")

def llamarfuncion():
  hora=datos()
  saludar_segun_hora(hora)


llamarfuncion()
```
