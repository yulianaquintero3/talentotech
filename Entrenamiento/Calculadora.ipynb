```python
import math

def mostrar_titulo():

  print('Ejemplo Calculadora: ')


def mostrar_menu():
    print("MENU:")
    print("1: Sumar (+)")
    print("2: Restar (-)")
    print("3: Multiplicar (*)")
    print("4: Dividir (/)")
    print("5: Logaritmo (log)")
    print("6: Coseno (cos)")
    print("7. Potencia (**)")
    print("8. Radicación ")
    print("9. Seno (sen)")
    print("10. Tangente (tan)")
    print("11. Salir")

def ingresar_datos():
  opcion=input("Ingrese la opción deseada del menú: ")
  return opcion


def calcular_suma(numero_uno,numero_dos):
  suma=numero_uno+numero_dos
  return suma

def calcular_resta(numero_uno,numero_dos):
  return numero_uno-numero_dos

def calcular_multiplicacion(numero_uno,numero_dos):
  return numero_uno*numero_dos

def calcular_division(numero_uno,numero_dos):
  return numero_uno/numero_dos

def calcular_logaritmo(numero_uno,numero_dos):
  logaritmo=math.log(numero_uno,numero_dos)
  return logaritmo

def calcular_coseno(angulo_radianes):
  grados = math.radians(angulo_radianes)
  coseno=math.cos(grados)
  return coseno

def calcular_potencia(numero_uno,numero_dos):
  return numero_uno**numero_dos

def calcular_raiz(numero_uno,numero_dos):
  return numero_uno**(1/numero_dos)

def calcular_seno(angulo_radianes):
  grados = math.radians(angulo_radianes)
  return math.sin(grados)

def calcular_tangente(angulo_radianes):
  grados = math.radians(angulo_radianes)
  return math.tan(grados)

def procesar_y_mostrar_operacion(opcion):
  if opcion=='1' or opcion=='+' or opcion=='Sumar':
    numero_uno=float(input('Ingrese el primer número para realizar la suma: '))
    numero_dos=float(input('Ingrese el segundo número para realizar la suma: '))
    resultado=calcular_suma(numero_uno,numero_dos)
    print('La suma es:',resultado)

  if opcion=='2'or opcion=='-' or opcion=='Restar':
    numero_uno=float(input('Ingrese el primer número para realizar la resta: '))
    numero_dos=float(input('Ingrese el segundo número para realizar la resta: '))
    resultado=calcular_resta(numero_uno,numero_dos)
    print("La resta es:",resultado)

  if opcion=='3'or opcion=='*' or opcion=='Multiplicar':
    numero_uno=float(input('Ingrese el primer número para realizar la multiplicación: '))
    numero_dos=float(input('Ingrese el segundo número para realizar la multiplicación: '))
    resultado=calcular_multiplicacion(numero_uno,numero_dos)
    print("El resultado es:",resultado)

  if opcion=='4'or opcion=='/' or opcion=='Dividir':
    numero_uno=float(input('Ingrese el numerador  para realizar la división: '))
    numero_dos=float(input('Ingrese el denominador para realizar la división: '))
    #VALIDANDO EL DENOMINADOR (debe ser diferente de cero)
    if numero_dos!=0:
      resultado=calcular_division(numero_uno,numero_dos)
      print("El resultado es:",resultado)
    else:
      print("¡No es posible dividir entre cero!")

  if opcion=='5'or opcion=='log' or opcion=='Logaritmo':
    numero_uno=float(input('Ingrese el número para hallar el logaritmo: '))
    numero_dos=float(input('Ingrese la base del logaritmo: '))
    resultado1=calcular_logaritmo(numero_uno,numero_dos)
    print("El resultado es:",resultado1)

  if opcion=='6'or opcion=='cos' or opcion=='Coseno':
    numero_uno=float(input('Ingrese el valor de un ángulo en grados para hallar el coseno: '))
    resultado1=calcular_coseno(numero_uno)
    print("El resultado es:",resultado1)

  if opcion=='7'or opcion=='**' or opcion=='Potencia':
    numero_uno=float(input('Ingrese el número base para la potencia: '))
    numero_dos=float(input('Ingrese exponente de la potencia: '))
    resultado=calcular_potencia(numero_uno,numero_dos)
    print("El resultado es:",resultado)

  if opcion=='8' or opcion=='Radicación ':
    numero_uno=float(input('Ingrese el número radicando para la raiz: '))
    numero_dos=float(input('Ingrese el índice de la raiz: '))
    resultado1=calcular_raiz(numero_uno,numero_dos)
    print("El resultado es:",resultado1)

  if opcion=='9'or opcion=='sen' or opcion=='Seno':
    numero_uno=float(input('Ingrese el valor de un ángulo en grados para hallar el seno: '))
    resultado1=calcular_seno(numero_uno)
    print("El resultado es:",resultado1)

  if opcion=='10'or opcion=='tan' or opcion=='Tangente':
    numero_uno=float(input('Ingrese el valor de un ángulo en grados para hallar la tangente: '))
    resultado1=calcular_tangente(numero_uno)
    print("El resultado es:",resultado1)

def llamar_funciones():
  mostrar_titulo()
  while True:
    print()
    mostrar_menu()
    opcion=ingresar_datos()
    procesar_y_mostrar_operacion(opcion)
    print('------------------------------------------------')
    if opcion=='11' or opcion.lower()=='salir':
      print("¡Hasta luego!")
      break



#EJECUCION DEL PROGRAMA:
llamar_funciones()
```
